```python
import cv2
from collections import deque
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from roidetection import get_next_rois, draw_rois
from scraper import crop_im

field_rois = np.array([[136,  79, 200, 361], [624,  79, 199, 361]])
next_rois = get_next_rois(field_rois)
```

```python
im = Image.open('test-images/amitie_witch.png').convert('RGB').resize((960, 540))
im = np.array(im)
roi_im = draw_rois(im, [field_rois, next_rois])
plt.imshow(im)
plt.show()
```

```python
p1_next = crop_im(im, next_rois[0])
plt.imshow(p1_next)
```

```python
roi = next_rois[0]

top_roi = np.copy(roi)
top_roi[2] = roi[2] * 0.5
top_roi[3] = roi[3] * 0.1
top_roi[0] = roi[0] + roi[2] * 0.08
top_roi[1] = roi[1] + roi[3] * 0.05

top = crop_im(im, top_roi)
plt.imshow(top)
```

```python
bot_roi = np.copy(roi)
bot_roi[2] = roi[2] * 0.5
bot_roi[3] = roi[3] * 0.1
bot_roi[0] = roi[0] + roi[2] * 0.35
bot_roi[1] = roi[1] + roi[3] * 0.52

bot = crop_im(im, bot_roi)
plt.imshow(bot)
```
